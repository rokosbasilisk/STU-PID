```python
import pandas as pd

# Load the JSONL file into a DataFrame
baseline_df = pd.read_json('baseline.jsonl', lines=True)
seal_df = pd.read_json('steering_coef1.jsonl', lines=True)
```

```python
baseline_df.head()
```

```python
seal_df.info()
```

```python
# Perform the merge on the specified key columns
merged_df = baseline_df.merge(
    seal_df,
    on=['prompt', 'problem', 'answer', 'solution'],
    suffixes=('_baseline', '_seal')
)

# Rename the model_generation columns
merged_df = merged_df.rename(columns={
    'model_generation_baseline': 'baseline_generation',
    'model_generation_seal': 'seal_generation'
})
```

```python
merged_df.to_csv('seal_results.csv')
```
